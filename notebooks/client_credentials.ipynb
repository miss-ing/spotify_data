```python
from dotenv import load_dotenv
import os
from pathlib import Path
from spot.web_api import get_client_credentials, get_access_token

env_file = Path(__file__).parent / "my_client_credentials.env"

load_dotenv(dotenv_path=env_file)


client_id = os.getenv('client_id')
client_secret = os.getenv("client_secret")

print(client_id, client_secret)

client_credentials = get_client_credentials(env_file )
print(client_credentials)

print (get_access_token(client_credentials))
```
