```python
track_info = pd.read_json(output_dir/"track_uri_info.json")

#delete columns 
track_info = track_info[['id','name','album','artists','duration_ms', 'explicit','disc_number','track_number','external_ids']]


#create M:N track:artist df
tracks_artists = track_info[['id','artists']].copy()
tracks_artists=tracks_artists.rename(columns={
    'id':'track_id',
    'artists':'artist_id'})
#get ids from artists object and normalise
tracks_artists['artist_id'] = tracks_artists['artist_id'].apply(
    lambda artists: [artist.get('id') for artist in artists])
tracks_artists = tracks_artists.explode('artist_id')


#create tracks table 
track_info = track_info[['id','name','album','artists','duration_ms','explicit','disc_number','track_number','external_ids']]
track_info  = track_info.rename(columns = {
    'id':'track_id', 
    'album': 'album_id', 
    'name':'track_name', 
    'artists':'primary_artist_id',
    'external_ids':'isrc_id'})
#extract ids from objects
track_info['primary_artist_id'] = track_info['primary_artist_id'].apply(lambda x: x[0]['id'])
track_info['album_id'] = track_info['album_id'].str['id']
#turn external_id into isrc if there is one 
track_info['isrc_id'] =track_info['isrc_id'].apply(
    lambda d: d.get('isrc') 
    if isinstance(d,dict) and 'isrc' in d 
    else None)
```
